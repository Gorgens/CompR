```
import pandas as pd
```

```
url="https://raw.githubusercontent.com/Gorgens/treelab/master/manejo/inventarioContinuo.csv"
inv=pd.read_csv(url)
inv.head()
```

```
inv2012 = inv.loc[:,['transect', 'tree', 'scientific.name', 'DBH.2012']]
inv2012.head()
```

```
inv2014 = inv.loc[:,['transect', 'tree', 'scientific.name', 'DBH.2014']]
inv2014 = inv2014[pd.notna(inv2014['DBH.2014'])]
# inv2014.shape
inv2014.head()
```

```
tree2012 = inv2012.pivot_table(values='tree', index='transect', aggfunc='count')
tree2014 = inv2014.pivot_table(values='tree', index='transect', aggfunc='count')
tree2012.join(tree2014, lsuffix='_12', rsuffix='_14')
```

$$S_j = \frac{c}{(a+b-c)}$$

```
especies2012 = inv2012.loc[pd.notna(inv2012['scientific.name']),'scientific.name'].sort_values().unique()
especies2014 = inv2014.loc[pd.notna(inv2014['scientific.name']),'scientific.name'].sort_values().unique()

# print(especies2012)
# print(especies2014)

a = len(especies2012)
print('Número de espécies em A:', len(especies2012))

b = len(especies2014)
print('Número de espécies em B:', len(especies2014))

c = len(set(especies2012).intersection(set(especies2014)))
print('Espécies comuns:', len(set(especies2012).intersection(set(especies2014))))

print('Similaridade:', c/(a+b-c))
```
