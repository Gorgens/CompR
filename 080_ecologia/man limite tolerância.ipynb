# Limite de tolerância

Encontramos vida em quase todos os lugares na Terra, mas ela não é distribuída uniformemente no planeta. Espécies diferentes são encontradas em diferentes áreas; algumas espécies se sobrepõem e outras não. Cada espécie tem um conjunto de condições ambientais dentro das quais consegue sobreviver e se reproduzir.

```
import numpy as np
import pandas as pd
import altair as alt
```

# Perguntas direcionadoras

1. Qual espécie tem um limite mais restritivo?
2. Estas espécies podem coexisitir?


```
temperaturaMedicoesPop1 = pd.DataFrame({'Temp': np.random.normal(loc=30.0, scale=3.0, size=300),
                                        'Populacao': ['pop1'] * 300})

temperaturaMedicoesPop2 = pd.DataFrame({'Temp': np.random.normal(loc=21.0, scale=9.0, size=300),
                                       'Populacao': ['pop2'] * 300})

populacao = pd.concat([temperaturaMedicoesPop1, temperaturaMedicoesPop2])

alt.Chart(populacao).mark_bar(
    opacity=0.5
).encode(
    alt.X("Temp:Q", bin=alt.Bin(maxbins=100)),
    alt.Y('count()', stack=None),
    alt.Color('Populacao:N')
)
```

```
temperaturaMedicoesPop1 = pd.DataFrame({'Temp': np.random.normal(loc=15.0, scale=1.5, size=300),
                                        'Populacao': ['pop1'] * 300})

temperaturaMedicoesPop2 = pd.DataFrame({'Temp': np.random.normal(loc=26.0, scale=1.5, size=300),
                                       'Populacao': ['pop2'] * 300})

populacao2 = pd.concat([temperaturaMedicoesPop1, temperaturaMedicoesPop2])

alt.Chart(populacao2).mark_bar(
    opacity=0.5
).encode(
    alt.X("Temp:Q", bin=alt.Bin(maxbins=100)),
    alt.Y('count()', stack=None),
    alt.Color('Populacao:N')
)
```
